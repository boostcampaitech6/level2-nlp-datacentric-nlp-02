```python
import pandas as pd
from konlpy.tag import Okt
from pykospacing import Spacing
```

```python
train = pd.read_csv('./train.csv')
```

```python
okt = Okt()
def okt_pre(text):
    prepro_words = []
    for word, tag in okt.pos(text):
        if tag not in ['Josa', 'Punctuation', 'Adverb']:
            prepro_words.append(word)
    prepro_sen = ' '.join(prepro_words)  # 각 단어 사이에 공백 추가
    return prepro_sen

def apply_okt(data):
    dataset = data.copy()
    dataset['text'] = dataset['text'].apply(okt_pre)
    return dataset
```

```python
train_v1 = apply_okt(train)
```

```python
train_v1.to_csv('./train_okt.csv', index=False)
```
