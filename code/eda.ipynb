# Distribution Check

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# og train
df = pd.read_csv('../data/train.csv')

label_summary = df.groupby('target').ID.count()

label = list(label_summary)
index = list(label_summary.index)

plt.figure()
plt.bar(index, label)
plt.title('train Target Count Summary')
plt.xlabel('count')
plt.ylabel('label')

for idx, val in enumerate(label_summary):
    plt.text(idx, val, str(val))

plt.show()
```
