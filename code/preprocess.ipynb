```python
import pandas as pd
import re
from hanspell.spell_checker import *


def hanspelling(texts):
    """
    hanspell로 맞춤법 검사를 실시합니다.
    """
    preprocessed_text = []
    for text in texts:
        text = check(text).checked
        if text:
            preprocessed_text.append(text)
    return preprocessed_text



def remove_characters(texts):
    """
    특수 기호를 제거합니다.
    """
    preprocessed_text = []
    for text in texts:
        # 정규 표현식을 사용하여 특수 기호를 제거
        cleaned_text = re.sub(r'[^\w\s]|_', '', text)
        preprocessed_text.append(cleaned_text)
    return preprocessed_text

def convert_uppercase(texts):
    """
    영어 알파벳을 대문자로 변환합니다.
    """
    uppercase_text = []
    for text in texts:
        # 영어 알파벳을 대문자로 변환
        converted_text = text.upper()
        uppercase_text.append(converted_text)
    return uppercase_text

def remove_startwords(texts):
    """
    텍스트가 '게시판' 또는 '그래픽' 등으로 시작하면 해당 단어를 제거합니다.
    """
    processed_texts = []
    for text in texts:
        # '게시판' 또는 '그래픽'으로 시작하는 경우 해당 단어를 제거
        cleaned_text = re.sub(r'^(게시판|그래픽|위클리|위클리 스마트|SNS 돋보기|1보|2보|종합|속보|카드뉴스|르포|연합이매진|연합뉴스)', '', text)
        processed_texts.append(cleaned_text)
    return processed_texts

def remove_endwords(texts):
    """
    텍스트가 '게시판' 또는 '그래픽' 등으로 끝나면 해당 단어를 제거합니다.
    """
    processed_texts = []
    for text in texts:
        # '게시판' 또는 '그래픽'으로 시작하는 경우 해당 단어를 제거
        cleaned_text = re.sub(r'(종합|1보|2보|속보)$', '', text)
        processed_texts.append(cleaned_text)
    return processed_texts



sents = hanspelling(df['text'])
sents = remove_characters(sents)
sents = convert_uppercase(sents)
sents = remove_startwords(sents)
sents = remove_endwords(sents)
df['text'] = sents
df.to_csv()
```
