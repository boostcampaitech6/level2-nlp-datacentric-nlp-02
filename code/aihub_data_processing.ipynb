# load raw data

```python
import json
import pandas as pd
```

```python
data = json.load(open('../data/TS_span_extraction.json'))
df = pd.json_normalize(data['data'])
```

```python
df_need = df[['doc_id', 'doc_title', 'doc_class.code', 'doc_source', 'doc_published']]
df_need.columns = ['ID', 'text', 'target', 'url', 'date']
df_need
```

```python
df_need['target'].unique()
```

```python
mask = df_need['target'].isin(['지역', '기타'])
df_need2 = df_need[~mask].reset_index(drop=True)
df_need2['target'].unique()
```

```python
df_need2['text'].value_counts()
```

```python
df_need2 = df_need2.drop_duplicates(subset='text')
df_need2['text'].value_counts()
```

```python
labels = {'IT과학':0, '경제':1, '사회':2, '문화':3, '국제':4, '스포츠':5, '정치':6}

for k, v in labels.items():
    df_need2['target'].replace(k, v, inplace=True)
    
df_need2
```

```python
out = df_need2.copy()
out = out.astype({'ID':object, 'text':object, 'target':int, 'url':object, 'date':object})
out
```

```python
out.to_csv('../data/loaded_aihub.csv', index=False)
```

# regex

```python
import re
import pandas as pd
```

```python
df1 = pd.read_csv('../data/loaded_aihub.csv')
df1
```

```python
text = df1['text'].to_list()
text
```

```python
# 테스트용
# \u2026 = 말줄임표(ellipsis)
# \u00B7 = interpunct
# \u2190-3 = arrows
remove_punc = re.compile(r'[^a-zA-Z0-9ㄱ-ㅎㅏ-ㅣ가-힣一-龥.%\u2026\u00B7\u2190\u2191\u2192\u2193]')
test = '추석 열차 승차권 예매 첫날 예매율 51.4%'

a = re.sub(pattern=remove_punc, repl=' ', string=test)

to_single_space = re.compile(r'\s+')
re.sub(pattern=to_single_space, repl=' ', string=a)
```

```python
# Remove punctuations
# bracket excluded ver.
remove_punc = re.compile(r'[^a-zA-Z0-9ㄱ-ㅎㅏ-ㅣ가-힣一-龥.%\u2026\u00B7\u2190\u2191\u2192\u2193]')
# bracket included ver.
# remove_punc = re.compile(r'[^a-zA-Z0-9ㄱ-ㅎㅏ-ㅣ가-힣一-龥.%\[\]\u2026\u00B7\u2190\u2191\u2192\u2193]')
# Multiple whitespace to single whitespace
to_single_space = re.compile(r'\s+')
# Remove special characters
special = re.compile(r'lt|gt')

new_txt = list(map(lambda x: re.sub(pattern=remove_punc, repl=' ', string=x).strip(), text))
new_txt1 = list(map(lambda x: re.sub(pattern=special, repl='', string=x), new_txt))
fin_txt = list(map(lambda x: re.sub(pattern=to_single_space, repl=' ', string=x).strip(), new_txt1))
fin_txt
```

```python
df1['text'] = fin_txt
df1.head(20)
```

```python
# bracket excluded ver.
df1.to_csv('../data/processed_aihub.csv', index=False)
# bracket included ver.
# df1.to_csv('../data/processed_aihub_bracket.csv', index=False)
```

## Uniform (Down sampling)

```python
cnt = ai_hub['target'].value_counts().sort_index()
cnt
```

```python
min = 2728
seed = 456

zero = ai_hub[ai_hub.target.isin(cnt[cnt.index == 0].index)].sample(min, random_state=seed)
one = ai_hub[ai_hub.target.isin(cnt[cnt.index == 1].index)].sample(min, random_state=seed)
two = ai_hub[ai_hub.target.isin(cnt[cnt.index == 2].index)].sample(min, random_state=seed)
three = ai_hub[ai_hub.target.isin(cnt[cnt.index == 3].index)].sample(min, random_state=seed)
four = ai_hub[ai_hub.target.isin(cnt[cnt.index == 4].index)].sample(min, random_state=seed)
five = ai_hub[ai_hub.target.isin(cnt[cnt.index == 5].index)].sample(min, random_state=seed)
six = ai_hub[ai_hub.target.isin(cnt[cnt.index == 6].index)].sample(min, random_state=seed)
six
```

```python
uniform_ai_hub = pd.concat([zero, one, two, three, four, five, six], axis=0).sample(frac=1, random_state=seed).reset_index(drop=True)
uniform_ai_hub
```

```python
uniform_ai_hub.to_csv('../data/uniform_ai_hub.csv')
```
